# 基本統計圖型

### plot():折線圖:matplotlib.pyplot.plot

https://github.com/MyFirstSecurity2020/20200825/blob/master/%E8%B3%87%E6%96%99%E7%A7%91%E5%AD%B8/2_Matplotlib%E5%AF%A6%E6%88%B0.md

```
import matplotlib.pyplot as plt
 
x  = [3, 6, 9, 12, 15, 18, 21, 24, 27]
y1 = [1, 4, 7, 6, 8, 6, 3, 5, 1]
y2 = [2, 5, 8, 7, 9, 7, 4, 6, 2]
plt.plot(x, y1, label="line L")
plt.plot(x, y2, label="line H")
plt.plot()

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Graph Example")
plt.legend()
plt.show()
```


### bar():長條圖|柱狀圖(Bar Chart):matplotlib.pyplot.bar

https://matplotlib.org/tutorials/introductory/pyplot.html


```
import matplotlib.pyplot as plt
names = ['gentleman', 'lady', 'Narcissistic', 'Connotative']  #紳士, 女士, 自戀的, 有內涵的 
values = [32, 40, 54, 18]
plt.bar(names, values)
plt.show()
```

### hist():直方圖(histogram):matplotlib.pyplot.hist

https://pythonspot.com/matplotlib-histogram/



```
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

x = [5,10,15,9,8,7,35,40,45,20,15,60,35,70,40,80,20,90,65,100]
num_bins = 5
n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.5)
plt.show()
```
